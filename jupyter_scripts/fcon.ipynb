```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import ScalarFormatter
```

```python
# Assuming your file is named 'data.txt'
df = pd.read_csv('../project/hype_model/0058418.txt', sep='\t',index_col=0)
```

```python
# Drop the first row of the dataframe
df = df.drop(df.index[0])
```

```python
# Convert index to datetime
df.index = pd.to_datetime(df.index)

# Trim the DataFrame to start at 2010-10-01
df = df[df.index >= '1988-10-01']
```

```python
df
```

```python
# Convert 'fcon' column to float
df['fcon'] = df['fcon'].astype(float)
```

```python
# Set the figure size
plt.figure(figsize=(12, 6))

# Plot fcon column against the index
plt.plot(df.index, df['fcon'])

# Set scientific notation for y-axis labels
plt.gca().yaxis.set_major_formatter(ScalarFormatter(useMathText=True))

# Calculate the range of the data
y_min = df['fcon'].min()
y_max = df['fcon'].max()

# Calculate the tick interval
tick_interval = (y_max - y_min) / 9  # 10 ticks including the minimum and maximum

# Set y-axis ticks
plt.yticks(np.arange(y_min, y_max + tick_interval, tick_interval))

# Set the labels and title
plt.xlabel('Date')
plt.ylabel('Contributing Area Fraction (-)')
plt.title('Contributing Area Fraction for 1989')

# Rotate the x-axis labels for better readability
plt.xticks(rotation=45)

# Show the plot
plt.show()
```
